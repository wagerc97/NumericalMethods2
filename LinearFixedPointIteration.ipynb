```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initial x
x0 = 3
```

```python
# loop
xk = x0
xklist = []
nstep = 20

# Define function phi1
phi1 = lambda xk: 1 + 2 / xk 

for i in range(nstep):
    xklist.append(xk)
    xk = phi1(xk)

plt.figure(figsize=(10,5))
plt.plot(np.arange(nstep), xklist, marker="o", color="blue")
plt.show()
print(f"Xk = {xk}")
```

```python
# loop
xk = x0
xklist = []
nstep = 10

# Define function phi1
phi2 = lambda xk: xk * xk - 2

for i in range(nstep):
    xklist.append(xk)
    xk = phi2(xk)
    
plt.figure(figsize=(10,5))
plt.loglog(np.ndarray(nstep), xklist, marker="o", color="red")
plt.show()
#print(f"Xk = {xk}")
```

#### Comparison
Both iterations used equivalent functions phi 1,2. However, the fixed point form of the function converged quickly, while the other result explodes. The non-linear term 'x' blows any error up.
